# Reproduce Extended Data Fig. 2: Changes in cellular composition.

They calculated frequencies of cell types in each sample from COVID-19 and control lungs, and compared the medians of the two groups to identify differences in frequency. Significance was assessed using a Wilcoxon rank-sum test.

x axes is all cell types.
y axes is cell frequency among all cells，

```python
#sample information can be found in metadata
metadata.columns
```

```python
#add 'cell_type_intermediate' from the metadata to ourdata
data["sample"]=metadata["biosample_id"]
data.head()
```

```python
fraction_data= data.iloc[1:,-3:]
fraction_data.head()
```

```python
#Total number of all cells in each sample
cells_per_sample=fraction_data["sample"].value_counts(sort=False).to_dict()
cells_per_sample
```

```python
#Total number of each cell type in each sample
boxdata=fraction_data[["sample","group","clusters"]].value_counts(sort=False).astype(int).reset_index(name='num_ty_sample')
boxdata=boxdata.rename(columns={'clusters':'cell_types'})
boxdata
```

```python
boxdata["total_cells"] = boxdata["sample"].map(cells_per_sample).astype(int)
boxdata
```

```python
#calculate frequency of each cell type in each sample
boxdata["frequency"]=boxdata.num_ty_sample / boxdata.total_cells
boxdata.head()
```

```python
plt.figure(figsize = (10,4))

ax = sns.boxplot(data=boxdata, x ="cell_types", y = "frequency", hue = "group", whis=1.5)

plt.xticks(rotation = 35, rotation_mode = 'anchor', ha = 'right')

plt.savefig('boxplot_1.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Mann Whitney U test and reproduce the figure with significance labeled on the plot

```python
import scipy.stats as sps
```

```python
#perform the Wilcoxon-Signed Rank Test
pvals = boxdata.groupby('cell_types').apply(lambda x: 
                                            sps.mannwhitneyu(
                                                              x[x.group.eq('COVID-19')].frequency, 
#loop through pvalues to retain two valid digits                                                              x[x.group.eq('Control')].frequency)[1])
formatted_pvals = []
for pvalue in pvals:
    formatted_pvals.append(f"p = {'%.2g' % pvalue}")
formatted_pvals
```

```python
#find a proper location to put the p-values
label_loc=boxdata.groupby("cell_types").frequency.max()
label_loc
```

```python
sns.set_style("white")

plt.figure(figsize=(10, 3))

ax = sns.boxplot(data=boxdata, x = 'cell_types', y = 'frequency', hue = 'group', whis=1.5,
                 palette={"Control":"#168B9F","COVID-19":"#A50945"}, saturation=1)

#change plot artists settings to make hollow boxs 
box_patches = [patch for patch in ax.patches if type(patch) == matplotlib.patches.PathPatch]
if len(box_patches) == 0: 
    box_patches = ax.artists
num_patches = len(box_patches)
lines_per_boxplot = len(ax.lines) // num_patches
for i, patch in enumerate(box_patches):
    # Set the linecolor on the patch to the facecolor, and set the facecolor to None
    col = patch.get_facecolor()
    patch.set_edgecolor(col)
    patch.set_facecolor('None')

    # Each box has associated Line2D objects (to make the whiskers, fliers, etc.)
    # Loop over them here, and use the same color as above
    for line in ax.lines[i * lines_per_boxplot: (i + 1) * lines_per_boxplot]:
        line.set_color(col)
        line.set_mfc(col)  # facecolor of fliers
        line.set_mec(col)  # edgecolor of fliers

sns.stripplot(data=boxdata, x = 'cell_types', y = 'frequency', hue = 'group', palette={"Control":"#168B9F","COVID-19":"#A50945"})       

plt.xlabel("Cell type")
plt.ylabel("Frequency among all cells")
plt.xticks(rotation = 90, rotation_mode = 'anchor', ha = 'right')   
sns.despine(left=True)

#loop through length of compare pairs to add pvalue on the plot
for i in range(len(label_loc)):
    lab_loc=label_loc[i]+0.05
    plt.text(x=1*i,y=lab_loc,s=formatted_pvals[i], ha='center', va='bottom',size='x-small')

#save figure
plt.savefig('boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```
