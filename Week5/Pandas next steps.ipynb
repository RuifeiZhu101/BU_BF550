```python
#Filtering
def my_filter(x):
    return x['new_cases'].mean()>1000.0

df.groupby('continent').filter(my_filter)
```

```python
#Apply some function on the data
def my_function(x):
    return x['new_cases'].mean()-x['new_cases'].std(),x['new_cases'].mean(),x['new_cases'].mean()+x['new_cases'].std()

df2=df.groupby('continent').apply(my_function)
df2
```
