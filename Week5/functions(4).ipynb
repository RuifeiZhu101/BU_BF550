# Growth of sequencies

Consider:

(a) $$x_n = x_{n-1} + 1/n$$
(b) $$x_n = x_{n-1} + c$$
(c) $$x_n = x_{n-1} + n$$
(d) $$x_n = c*x_{n-1} $$

How do they grow for large n?

```python

def oon(N, x0=0):
    n = np.arange(1,N+1)
    x = np.zeros(N)
    x[0] = x0 + 1
    for i in n[:-1]:
        x[i] = x[i-1] + 1/i
    return (n, x)

n, x = oon(1000)
```

```python
fig, ((axlin, axlogx), (axlogy, axloglog)) = plt.subplots(2, 2, figsize = (13, 8))
axlin.plot(n,x,'.')
axlin.set_title('testing for linear relationship: $y=ax+b$')
#axlin.set_xlabel('n')
axlin.set_ylabel('x')
axlogx.semilogx(n, x,'.')
axlogx.set_title('testing for logarithmic relationship: $y=a\ln(x)+b$')
#axlogx.set_xlabel('n')
#axlogx.set_ylabel('x');
axlogy.semilogy(n, x,'.')
axlogy.set_title('testing for exponential relationship: $y=ae^x+b$')
axlogy.set_xlabel('n')
axlogy.set_ylabel('x');
axloglog.loglog(n, x,'.')
axloglog.set_title('testing for power law relationship: $y=ax^b$')
axloglog.set_xlabel('n');
#axloglog.set_ylabel('x');
```

```python
# determining the coefficients
slope, intercept, r_value, p_value, std_err = sps.linregress(np.log(n), x)
plt.semilogx(n, x, 'k*')
plt.xlabel('n')
plt.ylabel('x')
plt.semilogx(n, np.log(n)*slope + intercept, 'r-')
plt.title('dependence x(n): x = {0:.3f}*log(n) + {1:.3f}'.format(slope, intercept));
```

# Derivatives

Let's have a function $y=f(x)$

Definition: The derivative of a function $f(x)$ at a point $x=x_0$,

$\frac{df(x=x_0)}{dx}$

is the slope of the tangent of that function at the point $x=x_0$.

In the language of mathematics: $\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}\frac{f(x_0+\Delta x)-f(x_0)}{\Delta x}$

Often the symbol for the derivative is a "prime": either $y'$ or $f'(x)$


```python
#We can get the numerical derivative of (numerical) function in NumPy using np.gradient
%matplotlib inline
plt.figure(figsize=(10,5))
x = np.linspace(-1,1,2000)
#plt.plot(x, x, 'b', label='$y=x$') # linear
plt.plot(x, x**2, 'r', label='$y=x^2$') # quadratic
plt.xlabel('x')
#plt.ylabel('y')
y=x**2
y_prime=np.gradient(y,x)
plt.plot(x, y_prime, 'b', label='$\\frac{dy}{dx}$')#Derivative
plt.legend(fontsize=15)
```

```python
#Let's calculate the derivative, and plot the tangent of the quadratic function in the point x=0.25:

x_0=x[1250]
y_0=x_0**2

x_0, y_0
```

```python
#The derivative, i.e. the slope at that point is:
y_prime[1250]
```

Remember, a line is a linear function $f(x)=A x+B$, where A is the slope.
Let's find B. Remember, $f(x_0)=Ax_0+B$, so $B=f(x_0)-Ax_0$.

```python
#Let's plot the function y=x**2, and a line with this slope through that point
B=y[1250]-y_prime[1250]*x[1250]

plt.figure(figsize=(10,5))
plt.plot(x, x**2, 'r', label='$y=x^2$') # quadratic
plt.xlabel('x')

plt.plot(x, y_prime[1250]*x+B, 'b', label='Tangent')#Derivative
plt.legend(fontsize=15)
#let's plot the point too
plt.plot(x_0, y_0, 'o')
```

What is the meanning of the formula:
$\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}\frac{f(x_0+\Delta x)-f(x_0)}{\Delta x}$?

If I create an array of $\Delta X$ values that goes to zero, and calculate the ratios for each,
that ratio will tend toward the derivative of the function.

```python
delta_x=np.linspace(0.05,0.75,7)
delta_x
```

Let's plot bunch of line that go trough the points $(x_0,f(x_0))$ and $(x_0+\Delta x, f(x_0+\Delta x))$

For each line, (for each $\Delta x$) we have two points that we know,

$f(x_0+\Delta x)=A(x_0+\Delta x)+B$ 

and

$f(x_0)=A(x_0)+B$

This is a system of two linear equations that we can easily solve:

$A=\frac{f(x_0+\Delta x)-f(x_0)}{\Delta x}$ 

and 

$B=f(x_0)-Ax_0$


```python
#Let's calculate A and B, and plot the lines for each element of delta_x
A=((x_0+delta_x)**2-x_0**2)/(delta_x)   
B=x_0**2-A*x_0
A,B
```

```python
plt.figure(figsize=(10,5))
x_line=np.linspace(0.0,1.0,1100)
plt.plot(x, x**2, 'r', label='$y=x^2$') # quadratic
plt.xlabel('x')

for i in range(A.size):
    plt.plot(x_line, A[i]*x_line+B[i], 'b', label='Tangent')#Derivative
#plt.legend(fontsize=15)
#let's plot the point too
plt.plot(x_0, y_0, 'o')
#Now we can plot the sections too
for dx in delta_x:
    plt.plot(x_0+dx,(x_0+dx)**2,'or')
```

Derivatives of some functions:

$y=a$ $\to$    $y'=0$

$y=x^n$ $\to$    $y'=nx^{n-1}$

$y=a^x$ $\to$    $y'=a^x\ln(a)$

Special case for the base of the natural logarithm:  $y=e^x$ $\to$    $y'=e^x$

Rules of derivation:

$y=f(x)g(x)$ $\to$ $y'=f'(x)g(x)+f(x)g'(x)$

$y=\frac{f(x)}{g(x)}$ $\to$ $y'=\frac{f'g-fg'}{g^2}$

Finally, composite function:

$y=f(g(x))$ $\to$ $y'=f'(g(x))g'(x)$


## Let's derive an example.

$f(x)=x^2$

Recall the formal definition of a derivative:

$\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}\frac{f(x_0+\Delta x)-f(x_0)}{\Delta x}$

So in this case:

$\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}\frac{(x_0+\Delta x)^2-x_0^2}{\Delta x}$

$\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}\frac{(x_0+2x_0\Delta x+\Delta x^2)-x_0^2}{\Delta x}$

$\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}\frac{2x_0\Delta x+\Delta x^2}{\Delta x}$

$\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}2x_0\frac{\Delta x}{\Delta x}+\lim_{\Delta x \to 0}\frac{\Delta x^2}{\Delta x}$

$\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}2x_0+\lim_{\Delta x \to 0}\Delta x$

$\frac{df(x=x_0)}{dx}=\lim_{\Delta x \to 0}2x_0+\lim_{\Delta x \to 0}\Delta x$

$\frac{df(x=x_0)}{dx}=2x_0$

Same result as:

$y=x^n$ $\to$    $y'=nx^{n-1}$


$y=x^2$ $\to$    $y'=2x^{2-1}=2x$ for $n=2$.

# Integrals

Inverse of derivative.

If:

$\frac{dF(x)}{dx}=f(x)$

then:

$F(x)=\int f(x)dx$ (Indefinite integral)

To find them out, look at a table of derivatives, and take the opposite direction.

Examples, $C$ is an arbitrary constant:

$\int x^n=\frac{x^{n+1}}{n+1} + C$ 

$\int e^x=e^x+C$

$\int \frac{1}{x}=\ln(x)+C$

Why the arbitrary constant? Integral of zero is a constant.



$f(x)=0$ $\to$ $\int f(x)=C$




Definite integrals, area under the curve of the function.

$\int_{a}^{b}f(x)dx$

```python
plt.figure(figsize=(10,5))
x=np.linspace(0.0,1.0,100)
plt.plot(x, x**2, 'r', label='$y=x^2$') # quadratic
plt.xlabel('x',fontsize=15)
plt.ylabel('$x^2$',fontsize=15)
plt.fill_between(
        x, 
        x**2, 
        where= (0.2 < x)&(x < 0.4),
        color= "b",
        alpha= 0.2)
plt.text(0.45,0.07,'$\int_{0.2}^{0.4}x^2dx$',fontsize=15)
plt.arrow(0.44,0.1,-0.1,-0.05,head_width=0.02)
plt.show()
```

The fundamental theorems of calculus:

If $F'(x)=f(x)$ then:

$\int_{a}^{b}f(x)dx=F(b)-F(a)$

So, to calculate the area under the $f(x)$ curve, find the indefinite integral $F(x)=\int f(x)dx$ first, the use the formula above. 

Also, if:

$F(x)=\int_{a}^{x}f(t)dt$

then:

$F'(x)=f(x)$


```python
#In our case above the integral is x^3/3 from x=0.2 to 0.4
a=0.2
b=0.4
integ=b**3/3-a**3/3
integ
```

What if we don't know the indefinite integral? We can integrate numerically.

We can split the area under the curve into N narrow rectangles wide $\Delta x$, and sum them up.

$\int_{a}^{b}f(x)dx=\lim_{\Delta x \to 0}\Sigma_{i=0}^{N}f(x_i)\Delta x$

$\int_{a}^{b}f(x)dx=\lim_{N \to \infty}\Sigma_{i=0}^{N}f(x_i)\Delta x$


```python
#Let's integrate x^2 from 0.2 to 0.4 numerically
plt.figure(figsize=(10,5))
x=np.linspace(0.0,0.5,100)
plt.plot(x, x**2, 'r', label='$y=x^2$') # quadratic
plt.xlabel('x',fontsize=15)
plt.ylabel('$x^2$',fontsize=15)
plt.fill_between(
        x, 
        x**2, 
        where= (0.2 < x)&(x < 0.4),
        color= "b",
        alpha= 0.2)
#plt.text(0.45,0.07,'$\int_{0.2}^{0.4}x^2dx$',fontsize=15)
#plt.arrow(0.44,0.1,-0.1,-0.05,head_width=0.02)

N=5
delta_x=0.2/N
x_tointegrate=np.linspace(0.2+delta_x/2,0.4-delta_x/2,N)
plt.bar(x_tointegrate, x_tointegrate**2, delta_x-0.001) # quadratic
plt.show()
```

```python
integral=delta_x*np.sum(x_tointegrate**2)
integral
#compare to 0.01866666666666667 from the formula
```
