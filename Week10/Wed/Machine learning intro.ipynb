### Machine Learning

Machine learning is a medley of methods to train models to fit multidimensional data. 

The data in contemporary research is typically extremely high dimensional. We work with hundreds and thousands of data points. Machine learning in large part is fitting models to this highly dimensional data. 


### Two types of Machine Learning:

###### Supervized 
(labels defined, methods to find their values)

###### Unsupervized 
(no labels defined, need to find labels, if any) 

##### Supervized examples:

Regressions

Classifications

##### Unsupervized examples:

Dimensionality reduction

Clustering 


Clustering can be supervized or not. If we know how many clusters we are looking for, that's supervized. If we have no idea how many clusters are there, if any, thet's unsupervized. 

Supervized is typically a simple comuptational algorithm. 

Unsupervized tends toward AI. 


Example: Let's have a group of cats and dogs mixed. 

Supervized learning is knowing what a cat and what a dog is, and finding which and separating the cats from the dogs in the group. 

Unsupervized learning is knowing nothing about cats and dogs. It is finding out that there are two types of animals in the group, and giving them a label, names: cats and dogs.


Example: Supervized grouping the data in 3 clusters. Unsupervized finding if the data ig grouped in clusters at all. 






#### Training and test sets in ML

Training set: The known data to train the model. 

Test set: Also known, but used to test how well is the model doing. 

###### The goal is to make predictions

Two types of predictions:

    1. Interpolations: predicting values between datapoints. 
    
    2. Extrapolations: predicting values outside of the data domain, outside of the exptremes.



```python
# Example interpolation

x=np.linspace(0,1,20)
y=x**2+np.random.normal(scale=0.1,size=20)

plt.scatter(x,y)
```

```python
#Let's fit the data with a 3 and 4 degree polynomials
#Then calculate y(0.5)

y3_fit=np.polynomial.polynomial.Polynomial.fit(x,y,deg=3)
y4_fit=np.polynomial.polynomial.Polynomial.fit(x,y,deg=4)

plt.scatter(x,y, label="Data")

x_fit=np.linspace(0,1,1000)
plt.plot(x_fit,y3_fit(x_fit),color='r',label='Fit n=3')
plt.plot(x_fit,y4_fit(x_fit),color='g',label='Fit n=4')
plt.scatter(0.5,y3_fit(0.5),color='r')
plt.scatter(0.5,y4_fit(0.5),color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

print("The n=3 fit gives y(0.5)=",y3_fit(0.5))
print("The n=4 fit gives y(0.5)=",y4_fit(0.5))
print("The values are close!")
print("These are interpolated values.")
```

```python
#let's do the same but extrapolate to x=-0.5

y3_fit=np.polynomial.polynomial.Polynomial.fit(x,y,deg=3)
y4_fit=np.polynomial.polynomial.Polynomial.fit(x,y,deg=4)

plt.scatter(x,y, label="Data")

x_fit=np.linspace(-1,1,1000)
plt.plot(x_fit,y3_fit(x_fit),color='r',label='Fit n=3')
plt.plot(x_fit,y4_fit(x_fit),color='g',label='Fit n=4')

plt.scatter(-0.5,y3_fit(-0.5),color='r')
plt.scatter(-0.5,y4_fit(-0.5),color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

print("The n=3 fit gives y(0.5)=",y3_fit(-0.5))
print("The n=4 fit gives y(0.5)=",y4_fit(-0.5))
print("The values are far apart!")
print("These are extrapolated values.")
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.neural_network import MLPClassifier
# Load data
X, y = fetch_openml("mnist_784", version=1, return_X_y=True)

# Normalize intensity of images to make it in the range [0,1] since 255 is the max (white).
X = X / 255.0

```

```python
print("The shape of X is ",X.shape)
X
```

```python
Y=np.array(X.iloc[1000]) #Try 800
Y=np.reshape(Y,(28,28))
plt.imshow(Y)
```

```python
# Split the data into train/test sets
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]
classifier = MLPClassifier(
    hidden_layer_sizes=(50,20,10),
    max_iter=100,
    alpha=1e-4,
    solver="sgd",
    verbose=10,
    random_state=1,
    learning_rate_init=0.1,)
# fit the model on the training data
classifier.fit(X_train, y_train)
```

```python
fig, axes = plt.subplots(1, 1)
axes.plot(classifier.loss_curve_, 'o-')
axes.set_xlabel("number of iteration")
axes.set_ylabel("loss")
plt.show()
```

```python
#Let's test it

Y_array=np.array(X_test.iloc[500]) #Try 800
Y=np.reshape(Y_array,(28,28))
plt.imshow(Y)
```

```python
y=classifier.predict(X_test)
y
```
