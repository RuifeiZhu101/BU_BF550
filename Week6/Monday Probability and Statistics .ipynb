# Random variables

To define random variables, one needs a set of possible event and the assignment of their probabilities. This needs to be consitent:

$$P(A \;\;or\;\; B) = P(A) + P(B)$$ for mutually exclusive events $A$ and $B$,

and the sum of the probabilities over all mutually exclusive possibilities gives 1. 

For example, consider the result of a throw of a dice. There are six elementary events: 1, 2, 3, 4, 5, 6. Each with probability of 1/6. A composite event is say an even number. This is realized by any of the following elementary events: 2, 4, 6. The probability of an even number is then 1/6 + 1/6 + 1/6 = 1/2.

One way to describe a random variable is through a table of elementary outcomes and their probabilities:

|$x_i$| 1 | 2 | 3 | 4 | 5 | 6 |
|-|-|-|-|-|-|-|
|$p_i$| 1/6 | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 |

So the distribution will never actually be exactly uniform. 

Number of obtained 1s is called frequency $N_1$. 
If the total number of trials is $N$, the relative frequency is $\frac{N_1}{N}$.

But how do we define the probability? In this case the probability (density) to get the side with number $1$ is the relative frequency in the limit of extremely large number of trials.

$P(1)=\lim_{N\to \infty} \frac{N_1}{N}$

The 'Law of large numbers' states that measures of probability, the means etc., tend to the theoretical ones only in the limits of a large number of measurements, trials.   

But what if we have a limited set of trials. What if we have a finite population, for example.

Let's consider a class of N students. 

$N_{tot}$ -total number of students

$N_M$ -number of males

$N_F$ -number of females

But, we can have a different classification

$N_B$ -number of biology majors

$N_{CS}$ -number of computer science majors

Let's choose a student at random. What is the probability that the student is female?

In this case, it is super simple:

$P(F)=\frac{N_F}{N_{tot}}$

Same for biology major:

$P(B)=\frac{N_B}{N_{tot}}$


# Statistics vs. probability

So far we talked about probability. What is statistics? How do they differ?

What if we cannot measure all of the population? 

In that case, we can estimate the probability by taking a finite sample.

$N^s_{tot}$ -number of students in a random sample

$N^s_M$ -number of males in the sample

$N^s_F$ -number of females in the sample

Now we can only $estimate$ the probability:

$P(M)=\frac{N^s_M}{N^s_{tot}}$

Statistics is the science of estimating the probabilities, means and other population measures, from finite samples. Statistics is the art of extracting information from data.



How confident are we about the estimates? 

$P(M)=\frac{N^s_M}{N^s_{tot}}$

The estimation of measures (statistics), and the confidence levels is what Statistics is all about.

Some basic terminology and notation:

Population is a collection of individuals. (Can be trials, measurements, runs etc. as with the dice)

$\bf{Variables}$:

    Quantitative (e.g. student_height=186cm)
    
    Qualitative (e.g. female)

Population $\bf{parameters}$:
    
For example population mean $\mu$ or $<x>$, if $x$ is the stochastic variable.

Sample $\bf{statistics}$:

For example sample mean $\bar{x}$, if $x$ is the stochastic variable.

$\bf{Errors}$:

Sampling errors: $<x>-\bar{x}$. These are OK, this is exactly what Statistics is about.

Non-sampling errors: Bad sampling, bad experiments. 


# Probability

How do we assign probabilities?

    1. Intuition, assumptions. Example: Assume that a coin is fair, assign probabilities of 0.5 to each side. 
    
    2. Estimate from relative frequencies. 
    
    3. If don't know anything, the assumption typically is of equaly likely outcomes. (Just like a case of dice) 
    

Two schools of probability: $\bf{Frequentist}$ and $\bf{Bayesian}$

Frequentist: Only measurements and estimates make sense.

Bayesian: Estimate our belief of outcome, based on an asuumption ($\bf{Prior}$).

https://en.wikipedia.org/wiki/Thomas_Bayes



Kolmogorov showed that the two approaches are equivalent. No need to fight.
Kolmogorov formulated the axioms of probability theory.  

https://en.wikipedia.org/wiki/Andrey_Kolmogorov



In probability we typically talk about $\bf{events}$. 

For exampe: Get heads when flipping a coin. 

Probability of the event is given by the law of large numbers, or by knowing the population. 

$A$ -event

$P(A)=\lim_{N\to \infty} \frac{N_A}{N}$

$P(A)\geq 0$ (First axiom)

Complementary event $A^c$. 

$P(A)+P(A^c)=1$ (Second axiom)

What if we have infinite number of (mutually exclusive) events? 

$P(\cup_{i=1}^{\infty} A_i)=\Sigma_{i=1}^{\infty}P(A_i)$ (Third axiom)

```python
#Sample space, Venn diagram
#%pip install matplotlib_venn
import matplotlib_venn as venn
A=[1,2,3]
B=[1,2,3,4,5,6]
A_c=[4,5,6]
venn.venn2([set(A),set(B)])
```

$B$ is called the $\bf{sample}$ $\bf{space}$.

Obviously $B=A\cup A^c$

So: $P(A)+P(A^c)=P(B)=1$

or:

$P(A^c)=1-P(A)$

$\bf{Independent}$ $\bf{ events}$

$P(A\ and\ B)=P(A)P(B)$ 

Example, throwing dice. Let's throw two of them at the same time. (Or two throws of a single one)
One throw has nothing to do with the other. They are independent.

What is the probability to throw a specific pair, say two ones?

How many pairs are there? In other words, what is $N$ - number of events, in this case?

$N=6*6$ 

So $P(1\ and\ 1)=P(1)P(1)$

Or: $N=N_aN_b$ 

so if 

$P(N_a=1)=\frac{1}{N_a}$

and

$P(N_b=1)=\frac{1}{N_b}$

then:

$P(N_a=1\ and\ N_b=1)=\frac{1}{N}=\frac{1}{N_aN_b}=\frac{1}{N_a}\frac{1}{N_b}=P(N_a=1)P(N_b=1)$



What if the events are not independent?

$\bf{Conditional}$ $\bf{probability}$

$P(A|B)$: probability $P(A)$ if $B$ happened (or $P(A)$ given $B$).

Example: 

    A class of 40 students. 20 males and 20 females. 5 biology majors. All biology majors are males. 
    Choose a student at random. What is the probability to be a biology major? 
    What is the probability to be a male? 
    
    If chosen student is a male, what is the pobability to be a biology major? 
    
    What is the probability for the student to be a male non-biology major?

$P(A\ and\ B)=P(A|B)P(B)$

$P(A\ and\ B)=P(B|A)P(A)$

In the case of the students:

$P(male)=0.5$

$P(biology)=0.125$

$P(biology|male)=0.25$

$P(biology\ and\ male)=P(biology|male)P(male)=0.25*0.5=0.125$

or 

$P(B|A)=\frac{P(A\ and\ B)}{P(A)}$

$P(A|B)=\frac{P(A\ and\ B)}{P(B)}$
